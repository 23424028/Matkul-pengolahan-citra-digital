```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('yuji.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_512 = cv2.resize(img, (512, 512))

img_256 = cv2.resize(img_512, (256, 256), interpolation=cv2.INTER_NEAREST)
img_128 = cv2.resize(img_512, (128, 128), interpolation=cv2.INTER_NEAREST)
img_64  = cv2.resize(img_512, (64, 64), interpolation=cv2.INTER_NEAREST)

# 3. Menampilkan Hasil
titles = ['Original 512x512', 'Resized 256x256', 'Resized 128x128', 'Resized 64x64']
images = [img_512, img_256, img_128, img_64]

plt.figure(figsize=(15, 5))
for i in range(4):
    plt.subplot(1, 4, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
