```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_gray = cv2.imread('yuji.jpeg', cv2.IMREAD_GRAYSCALE)

def quantize_image(img, bits):
    levels = 2**bits
    interval = 256 / levels
    res = (np.floor(img / interval) * interval).astype(np.uint8)
    return res

img_8bit = quantize_image(img_gray, 8)
img_4bit = quantize_image(img_gray, 4)
img_2bit = quantize_image(img_gray, 2)

titles = ['8-bit (256 levels)', '4-bit (16 levels)', '2-bit (4 levels)']
images = [img_8bit, img_4bit, img_2bit]

plt.figure(figsize=(15, 5))
for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
